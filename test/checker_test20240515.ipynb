```python
import requests
import base64
import hashlib
import secrets
from urllib.parse import urlencode

# 配置
client_id = 'a75ddc29587045ac916c74e96dc22959'
redirect_uri = 'http://localhost:8081'
authorization_endpoint = 'https://accounts.spotify.com/authorize'
token_endpoint = 'https://accounts.spotify.com/api/token'
scope = 'user-read-private user-read-email user-library-read user-follow-read user-read-recently-played'

# 生成 Code Verifier
code_verifier = ''.join(secrets.choice('ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789') for i in range(64))
code_verifier_bytes = code_verifier.encode()

# 計算 SHA-256 散列值
code_challenge = hashlib.sha256(code_verifier_bytes).digest()

# 將散列值進行 Base64 編碼
code_challenge_base64 = base64.urlsafe_b64encode(code_challenge).rstrip(b'=').decode()


# 構建授權 URL
auth_params = {
    'response_type': 'code',
    'client_id': client_id,
    'scope': scope,
    'code_challenge_method': 'S256',
    'code_challenge': code_challenge_base64,
    'redirect_uri': redirect_uri
}
auth_url = authorization_endpoint + '?' + urlencode(auth_params)

# 向 Spotify 授權服務器發送請求
print("請在瀏覽器中訪問以下 URL 以登錄並授權你的應用程序:")
print(auth_url)

# 解析返回的授權碼
authorization_code = input("請在重定向後的 URL 中提取授權碼: ")

# 用授權碼獲取訪問令牌
token_params = {
    'client_id': client_id,
    'grant_type': 'authorization_code',
    'code': authorization_code,
    'redirect_uri': redirect_uri,
    'code_verifier': code_verifier
}
response = requests.post(token_endpoint, data=token_params)
token_data = response.json()

# 保存和管理訪問令牌
access_token = token_data['access_token']
refresh_token = token_data['refresh_token']
expires_in = token_data['expires_in']

# 你可以根據需要保存 access_token 和 refresh_token，以便在以後的請求中使用它們
print("Access Token:", access_token)
print("Refresh Token:", refresh_token)
print("Expires in (seconds):", expires_in)
```

```python
# Get Followed Artists
# https://developer.spotify.com/documentation/web-api/reference/get-followed
response = requests.get('https://api.spotify.com/v1/me/following?type=artist', headers = headers)

pd.json_normalize(response.json()['artists']['items'])
```

```python
def fetch_all_artists(headers):
    base_url = 'https://api.spotify.com/v1/me/following?type=artist'

    all_artists = []
    next_url = base_url

    while next_url:
        response = requests.get(next_url, headers = headers)
        data = response.json()

        artists_on_page = data['artists']['items']
        all_artists.extend(artists_on_page)

        next_url = data["artists"]["next"]

    return all_artists

all_artists_data = fetch_all_artists(headers)

print("Total number of artists:", len(all_artists_data))
```

```python
pd.json_normalize(all_artists_data)
```
